```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import tensorflow as tf
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import matplotlib.pyplot as plt
import pickle
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from tensorflow.keras.layers import Conv2D, Input, MaxPooling2D, Dropout, concatenate, UpSampling2D
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, TensorBoard
from tensorflow.keras import backend as K
```

```python
in_window,out_window,epochs,task=240,24,50,'forecasting' #regression-complex
tf.random.set_seed(42)

model4 = tf.keras.Sequential([tf.keras.layers.Input(shape=(in_window, 14))])
model4.add(tf.keras.layers.Conv1D(128, 16, padding='same'))
model4.add(tf.keras.layers.MaxPool1D(
    pool_size=2,
    strides=2,
    padding='valid'))
model4.add(tf.keras.layers.BatchNormalization())
model4.add(tf.keras.layers.Conv1D(128, 8, padding='same'))
model4.add(tf.keras.layers.MaxPool1D(
    pool_size=2,
    strides=2,
    padding='valid'))
model4.add(tf.keras.layers.BatchNormalization())
# Replace the second convolutional layer with an LSTM layer
model4.add(tf.keras.layers.LSTM(128, return_sequences=True))  # You can adjust the number of units (e.g., 128) as needed
model4.add(tf.keras.layers.MaxPool1D(
    pool_size=2,
    strides=2,
    padding='valid'))
model4.add(tf.keras.layers.BatchNormalization())
# Replace the third convolutional layer with an LSTM layer
model4.add(tf.keras.layers.LSTM(128, return_sequences=True))  # You can adjust the number of units (e.g., 128) as needed
model4.add(tf.keras.layers.MaxPool1D(
    pool_size=2,
    strides=2,
    padding='valid'))
model4.add(tf.keras.layers.BatchNormalization())
model4.add(tf.keras.layers.Flatten())
model4.add(tf.keras.layers.Dropout(0.2))
model4.add(tf.keras.layers.Dense(out_window, activation=tf.keras.layers.LeakyReLU()))
model4.compile(loss=tf.keras.losses.MeanSquaredError(),
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
               metrics=tf.keras.metrics.MeanSquaredError())
model4.build()
model4.summary()


g='1h'
X = np.load(f"/content/drive/MyDrive/New Data (copy)/{g}/X_train_{g}.npy", allow_pickle=True)[:10000]
Y = np.load(f"/content/drive/MyDrive/New Data (copy)/{g}/Y_train_{g}.npy",allow_pickle=True)[:10000]

Y=Y[in_window:,0]*100
# Y=[(Y[i]>Y[i-1]).astype(int) for i in range(1,len(Y))]
X=X[:len(X)-out_window,:]
rolled_X,rolled_Y=[],[]
for i in range(len(X)-in_window+1):
    rolled_X+=[X[i:i+in_window,:]]
for i in range(len(Y)-out_window+1):
    rolled_Y+=[Y[i:i+out_window]]
if (len(rolled_X)==len(rolled_Y)):
    X_,Y_= [np.array(rolled_X),np.array(rolled_Y)]

X_train,X_test,Y_train,Y_test=train_test_split(X_,Y_,test_size=0.2,random_state=42)
X_train,X_validate,Y_train,Y_validate=train_test_split(X_train,Y_train,test_size=0.2,random_state=42)

def scheduler(epoch, lr):
  if epoch%3==0 and epoch>=3:
    return lr * 0.1
  else:
    return lr


history=model4.fit(X_train,Y_train,
                  epochs=epochs,
                  batch_size=64,
                  validation_data=(X_validate, Y_validate),
                  callbacks=[tf.keras.callbacks.LearningRateScheduler(scheduler),ReduceLROnPlateau(factor=0.1, patience=5, min_lr=0.00001, verbose=1)])
# model4.save(f"/content/drive/MyDrive/New Data (copy)/{g}_{task}-LSTM({in_window}to{out_window})_{epochs}_2CONV_2LSTM_BatchNorm_new.h5")

pred=model4.predict(X_test)
print('Test-MSE:',tf.keras.metrics.MeanSquaredError()(Y_test,pred))
origin=[]
output=[]
for i in range(len(pred)):
    y_hf = pred[i]
    y_og = Y_test[i]
    if i==0:
        hf_pred=y_hf
    else:
        hf_pred = 0.1*np.array(list(hf_pred[1:])+[0]) + np.array(list(0.9*np.array(y_hf[:-1]))+[y_hf[-1]])
    output+=[hf_pred[0]]
    origin+=[y_og[0]]
output+=[i for i in hf_pred[1:]]
origin+=[i for i in y_og[1:]]
orig,pred_ = np.array(origin),np.array(output)
plt.figure(figsize=(16,9))
k=20
index=np.random.randint(0,len(pred)-k+1)
plt.plot(pred_[index:index+k])
plt.plot(orig[index:index+k])
plt.legend([f"{g}_LSTM({in_window}to{out_window})_{epochs}_{task}-output",f"Original_{g}"])
plt.title('Performance Comparison')
plt.xlabel('Periods')
plt.ylabel('Values')
plt.show()
```
