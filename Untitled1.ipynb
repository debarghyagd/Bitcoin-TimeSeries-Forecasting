```python
class CustomExpLoss(tf.keras.losses.Loss):
    def __init__(self):
        super().__init__()

    def call(self, y_true, y_pred):
        abs_loss = tf.abs(tf.cast(y_true - y_pred, dtype=tf.float64))
        exp_loss = tf.exp(abs_loss)
        return tf.reduce_mean(tf.reduce_sum(exp_loss, axis=-1))
```

```python
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, BatchNormalization, LSTM, concatenate, Dense
from tensorflow.keras.models import Model

in_window,out_window,task=240,24,'forecasting' #regression-complex
tf.random.set_seed(42)

# Define input shape
input_shape = (240, 14)

# Input layer
input_layer = Input(shape=input_shape)

# Convolutional layers with MaxPooling1D and Batch Normalization
conv1 = Conv1D(filters=64, kernel_size=3, activation='relu')(input_layer)
conv1 = MaxPooling1D(pool_size=2)(conv1)
conv1 = BatchNormalization()(conv1)

conv2 = Conv1D(filters=128, kernel_size=3, activation='relu')(conv1)
conv2 = MaxPooling1D(pool_size=2)(conv2)
conv2 = BatchNormalization()(conv2)

# Flatten Convolutional layer output
flatten_conv = Flatten()(conv2)

# LSTM layers with Batch Normalization
lstm1 = LSTM(64, activation='relu', return_sequences=True)(input_layer)
lstm1 = BatchNormalization()(lstm1)
lstm2 = LSTM(64, activation='relu')(lstm1)
lstm2 = BatchNormalization()(lstm2)

# Concatenate the outputs after flattening
merged_layer = concatenate([flatten_conv, lstm2])

# Dense layers (ANN) with Batch Normalization
dense1 = Dense(256, activation='relu')(merged_layer)
dense1 = BatchNormalization()(dense1)
dense2 = Dense(128, activation='relu')(dense1)
dense2 = BatchNormalization()(dense2)

# Output layer
output_layer = Dense(24, activation='linear')(dense2)

# Create the model
model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss=CustomExpLoss(), metrics=['mse'])

# Print model summary
model.summary()
```
